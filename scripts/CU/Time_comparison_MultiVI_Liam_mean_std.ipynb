```python
import pandas as pd
from datetime import datetime
```

```python
df_bavae = pd.DataFrame(bavae).T
```

```python
df_multivi = pd.DataFrame(multivi).T
```

```python
df = pd.concat([df_bavae, df_multivi])
```

```python
df['time_seconds'] = df['time'].apply(lambda x: datetime.strptime(x,'%H:%M:%S').second + datetime.strptime(x,'%H:%M:%S').minute * 60 + datetime.strptime(x,'%H:%M:%S').hour * 3600)
```

```python
df['epochs'] = df['epochs'].astype(int)
```

```python
df['seconds_per_iteration'] = df['~time/it'].apply(lambda x: float(x.split('s/it')[0]))
```

```python
df
```

```python
df['model'] = df.index
```

```python
df['model'] = df['model'].apply(lambda x: 'MultiVI no filtering' if 'filt' in x else 'MultiVI default' if 'MultiVI' in x else 'BAVAE')
```
